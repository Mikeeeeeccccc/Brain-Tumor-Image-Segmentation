```python
from ultralytics import YOLO
%matplotlib inline
```

```python
model = YOLO('yolov8n.pt')

results = model.train(data='D:\\deep_learning_final\\yolo\\data.yaml', 
                      epochs=30,
                      batch=32, 
                      imgsz=640, 
                      device=0, 
                      )
```

```python
from ultralytics import YOLO

pre_model = YOLO("D:\\deep_learning_final\\yolo\\runs\detect\\train\\weights\\best.pt")

source="D:\\deep_learning_final\\yolo\\test\\images"

result = pre_model.predict(source, imgsz=640,save=True)

```

```python
import cv2
import os
import matplotlib.pyplot as plt


folder_path = 'test\labels'

# 處理推理結果
for res in result[:5]:
    # 原始圖片
    img = res.orig_img

    # 提取邊界框和標籤
    boxes = res.boxes.xyxy  # [[x_min, y_min, x_max, y_max, confidence], ...]
    confidences = res.boxes.conf  # Confidence scores
    class_ids = res.boxes.cls.int()  # Class IDs
    names = res.names  # Class name mapping

    file_name = res.path.split("\\")[-1]
    file_name = file_name.replace(".jpg", ".txt")
    print(file_name)
    ground_truth_path = os.path.join(folder_path, file_name)
    with open(ground_truth_path, 'r') as file:
        content = file.readlines()
        for line in content:
            # 使用空格分割每一行的内容
            ground_truth = line.strip().split()  # strip() 去除两端空白，split() 默认按空格分割

    # 繪製推理結果的邊界框
    for i, box in enumerate(boxes):
        x_min, y_min, x_max, y_max = map(int, box[:4])
        confidence = confidences[i]
        class_id = int(class_ids[i])
        label = f"{names[class_id]}: {confidence:.2f}"

        # 繪製推理框
        color = (0, 0, 255) if class_id == 1 else (255, 0, 0)  # Tumor: Red, Non-Tumor: Blue
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, 2)
        print('pred:', x_min, y_min, x_max, y_max )
        # 繪製標籤
        cv2.putText(img, label, (x_min, y_min + (y_max-y_min+20)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)


    # # 繪製 ground truth 邊界框
    
    img_height, img_width = img.shape[:2]
    
    x_min = float(ground_truth[3])
    y_min = float(ground_truth[4])
    x_max = float(ground_truth[7])
    y_max = float(ground_truth[8])

    # # 轉換為像素坐標
    x_min_real = int(x_min*img_width)
    y_min_real = int(y_min*img_height)
    x_max_real = int(x_max*img_width)
    y_max_real = int(y_max*img_height)

    print('ground truth:', x_min_real, y_min_real, x_max_real, y_max_real )

    # 設置顏色（通常使用不同顏色來區分ground truth和預測框）
    color = (0, 255, 0)  # Ground truth框：綠色
    label = f"GT: {names[class_id]}"

    # 繪製 ground truth 邊界框
    cv2.rectangle(img, (x_min_real, y_min_real), (x_max_real, y_max_real), color, 2)

    # 繪製標籤
    cv2.putText(img, label, (x_min_real, y_min_real - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

    # 將BGR轉換為RGB格式以便plt顯示
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 顯示圖片
    plt.figure(figsize=(10, 10))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(res.path.split("\\")[-1])  # 顯示圖片名稱
    plt.show()

```
